START

SET total = 0

INPUT number_of_items

FOR i = 1 TO number_of_items
    INPUT price
    total = total + price
END FOR

DISPLAY "Total amount =", total

INPUT payment

IF payment >= total THEN
    change = payment - total
    DISPLAY "Payment successful"
    DISPLAY "Change =", change
ELSE
    DISPLAY "Insufficient payment"
END IF

END
