```python
import numpy as np

print("Polynomial Root Finder")
print("1. Quadratic")
print("2. Cubic")
print("3. Quartic")

choice = input("Choose 1, 2, or 3: ")

if choice == "1":
    print("Equation: Ax^2 + Bx + C = 0")
    
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    
    coeffs = [A, B, C]

elif choice == "2":
    print("Equation: Ax^3 + Bx^2 + Cx + D = 0")
    
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    
    coeffs = [A, B, C, D]

elif choice == "3":
    print("Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")
    
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))
    
    coeffs = [A, B, C, D, E]

else:
    print("Invalid choice")
    exit()

roots = np.roots(coeffs)

print("Roots are:")
print(roots)
```
